```python
# Initializing and importing
from qiskit import IBMQ, Aer
from qiskit_ibm_provider import IBMProvider
IBMQ.save_account(open("ibm_token.txt","r").read())
IBMQ.load_account()
IBMQ.providers()
provider = IBMQ.get_provider("ibm-q")

import building_blocks
import runtime
```
